```python
origin = np.array([0,0])
a = np.array([1,1])
```

```python
s_fac = 2  # --> Scaling Factor
b = s_fac * a
```

```python
plt.plot((origin[0],a[0]), (origin[1],a[1]), color = 'b', linewidth = 5 )
plt.plot((origin[0],b[0]), (origin[1],b[1]), color = 'r')
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```

# 1.2

```python
def transformation_visualization(i_vec, scaling_factor):
    origin = np.array([0,0])
    a = i_vec
    s_fac = scaling_factor # --> Scaling Factor
    b = s_fac * a
    plt.plot((origin[0],a[0]), (origin[1],a[1]), color = 'b', linewidth = 5 )
    plt.plot((origin[0],b[0]), (origin[1],b[1]), color = 'r')
    plt.xlim(0,s_fac*1.25)
    plt.ylim(0,s_fac*1.25)
    plt.show()
```

```python
transformation_visualization(a,6)
```

```python
def rotation_vis(i_vec, rotation_degree):
    '''
    i_vec is a 2D numpy array and represent the input vector
    rotation_degree is a real number and represent the rotation in degree
    
    '''
    rotation_degree = rotation_degree
    rotation_matrix = np.ones((2,2))

    rotation_matrix[0,0] = math.cos(math.radians(rotation_degree))
    rotation_matrix[0,1] = -math.sin(math.radians(rotation_degree))
    rotation_matrix[1,0] = math.sin(math.radians(rotation_degree))
    rotation_matrix[1,1] = math.cos(math.radians(rotation_degree))
    origin = np.array([0,0])
    a = i_vec
    
    rotated_vector = np.matmul(rotation_matrix,a)
    
    plt.plot((origin[0],a[0]), (origin[1],a[1]), color = 'b', linewidth = 5 )
    plt.plot((origin[0],rotated_vector[0]), (origin[1],rotated_vector[1]), color = 'r')
    plt.show()
    
```

```python
test_input_vec = np.array([1,1])
degree = 30
rotation_vis(test_input_vec, degree)
```
